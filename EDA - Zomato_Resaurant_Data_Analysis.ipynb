## Exploratory Data Analysis and Visualization

Now we will try to plot some charts and find some stats from the data.

```python
#finding which currency is used in which country

final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
#Finding mean for aggregate rating
final_df['Aggregate rating'].mean()
```

```python
## Finding rating count by doing mutivariate analysis

ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

```python

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
## Finding aggregate ratingcount with the help of bivariate analysis and ploting it.

sns.histplot(figsize=(12,6))
sns.histplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

#### As we can see from above bar plot maximum number of aggregate rating ranges between  2.9 to 3.8

```python
sns.displot(final_df['Aggregate rating'])
```

```python
#Count plot - useful for plotting categorical features
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

#### As here we can see Orange is highest in number it also shows customers gives average rating mostly.

### Q1: Which country uses what cuurency ?

```python
## Here i am using groupby function to make group of countries and currency they are using.
## This function group the data on given condition.

final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
## Finding 10 best restaurants on the basis of average rating.

agreegate_rating_df = final_df.sort_values(by="Aggregate rating", ascending=False)
agreegate_rating_df.head(10)
```

```python
##  Finding which restaurants are better for two persons on the basis of average cost .
average_cost = final_df[['Country','Average Cost for two']].groupby(['Country','Average Cost for two']).size().reset_index()
average_cost_new = final_df.sort_values(by="Average Cost for two", ascending=True)
average_cost_new.head(10)
```

```python
## Finding how many restaurants have online delivery, table booking?

final_df[['Country','Has Table booking', 'Has Online delivery']].groupby(['Country','Has Table booking', 'Has Online delivery']).size().reset_index()
```

#### Finding top 3 countries having maximum number of order

```python
country_names = final_df.Country.value_counts().index
country_counts = final_df.Country.value_counts().values
plt.pie(country_counts[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```
